This adapts the idea to work with `minimal2.asm` which comes from our original "Why is the floating point algorithm weird" problem.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
magic_line = 136

def make_temp_asm_file(nops):
    with open("minimal2.asm") as f:
        with open("temp.asm", mode='w') as outf:
            for index, line in enumerate(f):
                outf.write(line)
                if index == magic_line - 2:
                    break
            for n in range(nops):
                outf.write("\tnop\n")
            # Skip `.p2align` line
            f.readline()
            for line in f:
                outf.write(line)
                
# Test
#make_temp_asm_file(2)
```

This version gives all the timings back as a list

```python
import subprocess
import os

def run_test():
    """run_test: assembles, links and runs temp.asm
    returns best and average time
    assumes temp.exe prints the time as the first thing on each line"""
    with open("temp.txt","w") as output:
        if subprocess.call("as temp.asm -o temp.o",stdout=output) != 0:
            raise Exception("Failed to assmeble")

    # Pause for a moment to wait for the filesystem to catchup
    # Stops a bug at least on my Windows laptop
    import time
    time.sleep(0.2)
    with open("temp.txt","w") as output:
        if subprocess.call("g++ temp.o -o temp.exe",stdout=output) != 0:
            raise Exception("Failed to link")

    with open("temp.txt","w") as output:
        if subprocess.call("temp",stdout=output) != 0:
            raise Exception("Failed to run")

    timings = []
    with open("temp.txt") as input:
        for index, line in enumerate(input):
            data = line.split()
            timings.append( float(data[0]) )
    
    os.remove("temp.asm")
    os.remove("temp.o")
    os.remove("temp.exe")
    os.remove("temp.txt")

    return timings
```

```python
results = []
for nops in range(100):
    make_temp_asm_file(nops)
    ti = run_test()
    print(nops,"-->",ti)
    results.append(ti)
```

Draw a scatter plot, which visually displays very well that, well, something weird is happening!

```python
x = []
y = []
for nops, times in enumerate(results):
    for t in times:
        x.append(nops)
        y.append(t)

plt.scatter(x,y)
```

Version which skips the lowest runs

```python
toskip = 5

x = []
y = []
for nops, times in enumerate(results):
    stimes = sorted(times)
    stimes = stimes[:-toskip]
    for t in stimes:
        x.append(nops)
        y.append(t)

plt.scatter(x,y)
```

```python
print("Average of best 15 runs:")
toskip = 5
for nops, times in enumerate(results):
    stimes = sorted(times)
    stimes = stimes[:-toskip]
    print(nops," --> ",sum(stimes)/15)
```
